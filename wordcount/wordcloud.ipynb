```python
#soruce for sample speeches: https://avalon.law.yale.edu/subject_menus/inaug.asp

#https://www.datacamp.com/tutorial/wordcloud-python

#By specifying r before the path, we tell Python to interpret it as a raw string. This way, it will not consider \ as an escape character.
# Source: https://www.delftstack.com/howto/python/python-open-file-in-different-directory/#use-the-raw-strings-to-open-files-in-other-directories-in-python
user_input = input("which file do you want to run? ")
file_name = user_input + ".txt"

# Avoding escapes strings with using \\ instead of \ 
#it is advisable to use \\ in place of \. It removes the error of interpreting \ as an escape character.
folder_path = "/home/ks/Documents/Programming/Python/Practice/wordcount/samples/"

fullpath = folder_path + file_name

with open(fullpath , encoding="utf8") as myfile:
    # getting the sample text file
    
    lines = myfile.read()   
```

```python
lines = lines.lower()
lines = lines.replace("--"," ")
lines = lines.replace("."," ")
lines = lines.replace(","," ")
lines = lines.replace(";"," ")
lines = lines.replace(":"," ")
lines = lines.replace("-"," ")
lines = lines.replace("“"," ")
lines = lines.replace("”","")
lines = lines.replace("’re","")
lines = lines.replace("’s","")
lines = lines.replace("n’t","")
lines = lines.replace("\n", '')
lines = lines.replace("(", '')
lines = lines.replace(")", '')

```

```python
print(type(lines))
```

```python
words = lines.split() # created a word list from imported text file (variable lines)
stopwords = ['will','us',"ca", "d","c","applause"]
#stopwords = ['will','us',"america","american","ca","americans", "d","c","applause"]   
for word in list(words):  # iterating on a copy since removing will mess things up
    if word in stopwords: 
        words.remove(word) #removing the words from stopwords list one by one

lines = " ".join(words) 

print(lines)
print(len(lines))
```

```python
words = lines.split()
print(type(words))
dict_word = {}
for word in words:
    if word in dict_word:
        dict_word[word] += 1
    else:
        dict_word[word] = 1


for key in list(dict_word.keys()):
    print(key, ":", dict_word[key])
```

```python
# Import package
import matplotlib.pyplot as plt
# Define a function to plot word cloud
def plot_cloud(wordcloud):
    # Set figure size
    plt.figure(figsize=(40, 30))
    # Display image
    plt.imshow(wordcloud) 
    # No axis details
    plt.axis("off");
```

```python
# Import package
from wordcloud import WordCloud, STOPWORDS
# Generate word cloud
wordcloud = WordCloud(width= 3000, height = 2000, random_state=1, background_color='white', colormap='Spectral', collocations=False, stopwords = STOPWORDS).generate(lines)
# Plot
plot_cloud(wordcloud)
```

```python
#
wordcloud.to_file(f"results/{user_input}.png")
print(f"{user_input}" +" " +"has been saved" )
```
