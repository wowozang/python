```python
import os

folder_name=os.getcwd()

print(folder_name)
```

```python
fullfoldername = (folder_name) + "/samples/"

sample_names= os.listdir(fullfoldername)
print(sample_names)
```

```python

file_path_all = []
for samples in sample_names:
    folder_path = fullfoldername
    fullpath = folder_path + samples
    file_path_all.append(fullpath)
   
print(file_path_all)   
```

```python
fulltext_all = [] 
for files in file_path_all:
     
    with open(files , encoding="utf8") as myfile:
        lines = myfile.read()
        fulltext_all.append(lines) #appending full text strings for each loop; in a list called fulltext list.

lines = " ".join(fulltext_all) #converting the full_textlist to a string variable called "lines"
#print(lines)
```

```python
#print(type(lines))
```

```python
lines = lines.lower()
lines = lines.replace("--"," ")
lines = lines.replace("."," ")
lines = lines.replace(","," ")
lines = lines.replace(";"," ")
lines = lines.replace(":"," ")
lines = lines.replace("-"," ")
lines = lines.replace("“"," ")
lines = lines.replace("”","")
lines = lines.replace("’re","")
lines = lines.replace("’s","")
lines = lines.replace("n’t","")
lines = lines.replace("\n", '')
lines = lines.replace("(", '')
lines = lines.replace(")", '')

```

```python
words = lines.split() # created a word list from imported text file (variable lines)
#stopwords = ['will','us',"ca", "d","c","applause"]
stopwords = ['will','us',"america","american","ca","americans", "d","c","applause","nation","must"]   
for word in list(words):  # iterating on a copy since removing will mess things up
    if word in stopwords: 
        words.remove(word) #removing the words from stopwords list one by one

lines = " ".join(words) 

print(len(lines))
```

```python
words = lines.split()
dict_word = {}
for word in words:
    if word in dict_word:
        dict_word[word] += 1
    else:
        dict_word[word] = 1


for key in list(dict_word.keys()):
    print(key, ":", dict_word[key])
```

```python
# Import package
import matplotlib.pyplot as plt
# Define a function to plot word cloud
def plot_cloud(wordcloud):
    # Set figure size
    plt.figure(figsize=(40, 30))
    # Display image
    plt.imshow(wordcloud) 
    # No axis details
    plt.axis("off");
```

```python
# Import package
from wordcloud import WordCloud, STOPWORDS
# Generate word cloud
wordcloud = WordCloud(width= 3000, height = 2000, random_state=1, background_color='white', colormap='Spectral', collocations=False, stopwords = STOPWORDS).generate(lines)
# Plot
plot_cloud(wordcloud)
```

```python
wordcloud.to_file("results/All.png")
print(f"all" +" " +"has been saved" )
```
